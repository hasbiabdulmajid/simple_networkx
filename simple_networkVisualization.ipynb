# Simple Network Visualization
Network visualization that represents transportation routes that connect major cities in West Java using networkx
<br> <br>

## Import Library

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Preparing Data

```python
G = nx.Graph()
```

### Add Node
Position is manually added based on the approximate position seen on google map

```python
G.add_node("Sukabumi",pos=(2,0))
G.add_node("Cianjur",pos=(3,2))
G.add_node("Bogor",pos=(1,3))
G.add_node("Bekasi",pos=(3,5))
G.add_node("Cikampek",pos=(4,4))
G.add_node("Subang",pos=(7,3))
G.add_node("Purwakarta",pos=(5,3))
G.add_node("Bandung",pos=(6,0))
G.add_node("Sumedang",pos=(8,1))
G.add_node("Indramayu",pos=(10,6))
G.add_node("Cirebon",pos=(12,3))
G.add_node("Tasikmalaya",pos=(10,-2))
G.add_node("Banjar",pos=(13,-3))
```

### Add Edge
Edge added when there is direct access between 2 city

```python
G.add_edge("Bekasi","Cikampek")
G.add_edge("Sukabumi","Cianjur")
G.add_edge("Bogor","Cianjur")
G.add_edge("Bogor","Sukabumi")
G.add_edge("Cikampek","Purwakarta")
G.add_edge("Cikampek","Subang")
G.add_edge("Cikampek","Indramayu")
G.add_edge("Purwakarta","Bandung")
G.add_edge("Purwakarta","Subang")
G.add_edge("Cianjur","Bandung")

G.add_edge("Bandung","Sumedang")

G.add_edge("Subang","Cirebon")
G.add_edge("Indramayu","Cirebon")
G.add_edge("Sumedang","Cirebon")
G.add_edge("Bandung","Tasikmalaya")
G.add_edge("Bandung","Subang")
G.add_edge("Tasikmalaya","Banjar")
G.add_edge("Tasikmalaya","Sumedang")
```

## Plot Data

```python
pos=nx.get_node_attributes(G,'pos')
```

### Normal Plot

```python
#G.nodes()
nx.draw(G,pos, with_labels=True)
plt.show()
```

### Adding More Customization

```python
# Draw Nodes
nx.draw_networkx_nodes(G, pos, node_color="tab:red")

# Draw Edge
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_edges(
    G,
    pos,
    width=8,
    alpha=0.5,
    edge_color="tab:red",
)

# Draw Label
pos_higher = {}
y_off = -0.6  # offset on the y axis

for k, v in pos.items():
    pos_higher[k] = (v[0], v[1]+y_off)


nx.draw_networkx_labels(G, pos_higher,font_size=10, font_color="blue")

plt.tight_layout()
plt.axis("off")
plt.show()
```

## Degree Centrality

```python
nx.degree_centrality(G)
```

```python
dc = nx.degree_centrality(G)
plt.bar(range(len(dc)), list(dc.values()), align='center')
plt.xticks(range(len(dc)), list(dc.keys()),rotation='vertical')
plt.title("Degree Centrality")


plt.show()
```

## Betweenness Centrality

```python
nx.betweenness_centrality(G)
```

```python
bc = nx.betweenness_centrality(G)
plt.bar(range(len(bc)), list(bc.values()), align='center')
plt.xticks(range(len(bc)), list(bc.keys()),rotation='vertical')
plt.title("Betweenness Centrality")


plt.show()
```

So, based on degree centrality and betweeness centrality calculation it can be concluded that **Bandung** is the most important city in West Java
